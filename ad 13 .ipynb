```python
df=sns.load_dataset('taxis')
df
```

```python
sns.barplot(x='payment',y='total',data=df,color='blue',estimator='std')#in estimator you  can take mean std sum..
plt.show()
plt.tight_layout()
```

```python
sns.histplot(y=df['fare'],color='yellow')# by x and y we can change the x axis and y axis direction
plt.show()
```

```python
sns.histplot(x=df['fare'],palette=['red','blue'],hue=df['payment'])# by palette we can use diff colors and we can use r,g shortcut instead if green red
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=True,kde=False)#histogram grafh
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=False,kde=True)#kde grafh
plt.show()
```

```python
sns.distplot(x=df['fare'],hist=True,kde=True)#kde grafh
plt.show()
```

```python
sns.boxplot(x=df['fare'],color='purple')
plt.show()
```
