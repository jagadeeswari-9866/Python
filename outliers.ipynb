```python
from scipy.stats  import zscore
import seaborn as sns
import numpy as np
data=sns.load_dataset('iris')
df=data.copy()
```

```python
df
```

```python
(4.9-df['sepal_length'].mean())/df['sepal_length'].std()
```

```python
z_scores = np.abs(zscore(df.drop('species',axis=1)))#to get these values by zscore method
z_scores
```

```python
non_outliers = (z_scores <= 3).all(axis=1)# wuthour the outlier removes the outlier
df_no_outliers = df[non_outliers]
```

```python
df_no_outliers
```

```python
(z_scores>3).all(axis=1)
```

```python
outliers = (z_scores > 3).any(axis=1)#to know the outlier 0r axis the outlier
outlier_rows = df[outliers]
outlier_rows
```

```python
from sklearn.ensemble import IsolationForest
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = sns.load_dataset("iris")
data
```

```python
from sklearn.preprocessing import  LabelEncoder
lb=LabelEncoder()
data['species']=lb.fit_transform(data['species'])
```

```python
data.head(3)
```

```python
clf = IsolationForest(random_state=293,contamination=0.1)
clf.fit(data)
```

```python
y_hat=clf.predict(data)#to predict data whitch are outliers and non outliers
y_hat
```

```python
np.where(y_hat==-1)#these are outliers according to IsolationForest
```
